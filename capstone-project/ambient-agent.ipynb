```python
# Initialisierung der LLMs für die Agenten
llm = OpenAI(temperature=0)

# 1. Ambient Automation - Routenoptimierung
def optimize_shipping_route(data):
    prompt = f"Analysiere die aktuellen Seeroutenbedingungen für {data} und finde die effizienteste Route. Berücksichtige Wetter, Hafenkapazität und Verzögerungen."
    response = llm.invoke(prompt)
    return f"Empfohlene Seeroute: {response}"

route_tool = Tool(
    name="Routenoptimierung",
    func=optimize_shipping_route,
    description="Berechnet optimale Seerouten basierend auf Wetter, Hafenverfügbarkeit und Verzögerungen"
)

# 2. Assistant Automation - Personaldisposition
def schedule_crew_shifts(data):
    prompt = f"Optimiere die Schichtplanung für Crewmitglieder auf einem Frachtschiff. Berücksichtige gesetzliche Ruhezeiten und Effizienz. Input: {data}"
    response = llm.invoke(prompt)
    return f"Optimierte Schichtplanung: {response}"

crew_tool = Tool(
    name="Personaldisposition",
    func=schedule_crew_shifts,
    description="Plant Crew-Schichten unter Berücksichtigung von Verfügbarkeit und Vorschriften"
)

# 3. Monitoring Automation - Nachhaltigkeitsüberwachung
def monitor_emissions(data):
    prompt = f"Überwache die CO2-Emissionen eines Frachtschiffs und schlage Maßnahmen zur Reduktion vor. Input: {data}"
    response = llm.invoke(prompt)
    return f"CO2-Emissionsbericht: {response}"

sustainability_tool = Tool(
    name="Nachhaltigkeitsmonitoring",
    func=monitor_emissions,
    description="Überwacht CO2-Emissionen und schlägt Reduktionsmaßnahmen vor"
)

# 4. Multi-Agent System - Predictive Maintenance & Flottenmanagement
def predict_ship_maintenance(data):
    prompt = f"Analysiere Sensordaten von Frachtschiffen und bestimme den Wartungsbedarf. Input: {data}"
    response = llm.invoke(prompt)
    return f"Wartungsvorhersage: {response}"

def coordinate_fleet(data):
    prompt = f"Optimiere die Flottenlogistik basierend auf Wartungsintervallen und Verzögerungen. Input: {data}"
    response = llm.invoke(prompt)
    return f"Flottenkoordination: {response}"

maintenance_tool = Tool(
    name="Predictive Maintenance",
    func=predict_ship_maintenance,
    description="Erkennt frühzeitig Wartungsbedarf von Frachtschiffen"
)

fleet_tool = Tool(
    name="Flottenkoordination",
    func=coordinate_fleet,
    description="Koordiniert Flottenbewegungen basierend auf Wartungsintervallen und Verzögerungen"
)

# Multi-Agenten-System mit verschiedenen Automatisierungen
agent_executor = initialize_agent(
    tools=[route_tool, crew_tool, sustainability_tool, maintenance_tool, fleet_tool],
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

# Beispiel-Demo für jede Automation
print("--- Routenoptimierung (Ambient) ---")
print(agent_executor.run("Finde die beste Seeroute von Südchina nach Europa basierend auf Wetter & Verzögerungen"))

print("--- Personaldisposition (Assistant) ---")
print(agent_executor.run("Plane die Schichten für 20 Crewmitglieder auf einer Transatlantik-Route"))

print("--- Nachhaltigkeitsmonitoring (Monitoring) ---")
print(agent_executor.run("Überwache die CO2-Emissionen der Flotte und schlage Optimierungsmaßnahmen vor"))

print("--- Multi-Agenten Predictive Maintenance & Flottenkoordination ---")
print(agent_executor.run("Optimiere die Wartungsplanung und Flottenbewegungen für die gesamte Tailwind-Schiffsflotte"))
```
